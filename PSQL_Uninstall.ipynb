# `Find PSQL in Terminal: get the version installed`

+ You may want to delete one version or the entire PSQL eco-system. Depends on your needs
    + In the Terminal: you can type to get version number

A. ) `postgres -V`

B. ) `locate bin/postgres` then take that result and do

`/usr/lib/postgresql/13.11/bin/postgres -V` (where 13.11 is an example of a version you may have)
  
C. ) If you have more than one install: `locate bin/psql`

`------------------------------------------------------`

# Location where `PSQL` should be installed: 3 main areas
+ `/Library/PostgreSQL`
+ `/Applications/PostgreSQL`
+ `Users & Groups`

+ *This is not all inclusive because depending on your installation we have a few pieces lines to run*

**`Your version will be different so put in place of 9.1,13.11,15, etc`**

If `uninstall-postgresql` does not work exchange for `uninstall-postgres`

1. ) `open /Library/PostgreSQL/9.1/uninstall-postgresql.app`

2. ) `open /Library/PostgreSQL/uninstall-postgresql.app`

3. ) `sudo /Library/PostgreSQL/13.11/uninstall-postgresql.app/Contents/MacOS/installbuilder.sh`
I was then prompted for a password

Now you can run this and see there is nothing there anymore: `cd /Library/PostgreSQL/`

# `Deleting Local Files & Folders:`

4. ) `sudo rm /etc/postgres-reg.ini`

5. ) `sudo rm -rf /Library/PostgreSQL`

6. ) `sudo rm -rf /usr/local/var/postgres`

# `Enter the Users & Groups:`

7. ) `Click Apple Icon Upper-Left -> System Preferences -> Users & Groups`

+ Look at the left side and there should be a PSQL User, you will need to unlock the pad lock with your password
+ Then delete this user by hitting the minus sign and then lock padlock and close out


# `Optional Deletion to Free Up Shared Memory:`

`sudo rm /etc/sysctl.conf`


`--------------------------------------------------------------------`

8. ) If you had or have Homebrew:
* `brew list`
* `brew uninstall postgresql@14` after the at symbol your version
    + **`Remove local Psql files from Homebrew:`**
        + `rm -rf /usr/local/var/postgres` 
        
        + `rm /usr/local/var/log/postgres.log`
        
        + `rm -f ~/.psqlrc ~/.psql_history`
